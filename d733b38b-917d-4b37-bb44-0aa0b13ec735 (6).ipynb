## –ú–æ–¥–µ–ª—å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, –≥–¥–µ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ ‚Äî —Ñ–∞–∫—Ç –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–∞ (churn) –≤ —Å–ª–µ–¥—É—é—â–µ–º –º–µ—Å—è—Ü–µ:

* –†–∞–∑–±–µ–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É —Ñ—É–Ω–∫—Ü–∏–µ–π train_test_split().
* –û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:
    * –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
    * —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.
* –û—Ü–µ–Ω–∏—Ç–µ –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?

–ù–µ –∑–∞–±—É–¥—å—Ç–µ —É–∫–∞–∑–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä random_state –ø—Ä–∏ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–∏ –≤—ã–±–æ—Ä–∫–∏ –∏ –∑–∞–¥–∞–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞.

–ü–æ–¥–∞–µ–º —Ä–∞–∑–º–µ—á–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ. C —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ–º –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –≤ –ø—Ä–æ–ø–æ—Ä—Ü–∏—è—Ö 80% –∫ 20%.

```python
# —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –Ω–∞—à–∏ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–º–∞—Ç—Ä–∏—Ü–∞ X) –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é (y)
X = df.drop('churn', axis = 1)
y = df['churn']

# —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify = y)
```

**–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è**

```python
# –æ–±—É—á–∏—Ç–µ StandartScaler –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
scaler = StandardScaler()
scaler.fit(X_train)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–π—Ç–µ –æ–±—É—á–∞—é—â–∏–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–µ –Ω–∞–±–æ—Ä—ã –¥–∞–Ω–Ω—ã—Ö
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
# –∑–∞–¥–∞–¥–∏–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
lr_model = LogisticRegression(random_state=0)
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å
lr_model.fit(X_train_st, y_train)
# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
```

**–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å**

```python
# –∑–∞–¥–∞–¥–∏–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) # –í–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
rf_model.fit(X_train, y_train)# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:, 1]
```

```python
# –∑–∞–¥–∞—ë–º –º–æ–¥–µ–ª—å k_means —Å —á–∏—Å–ª–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ 5 –∏ —Ñ–∏–∫—Å–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ random_state
km = KMeans(n_clusters=5, random_state=0)
# –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏–π (–∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç –∏–º –Ω–æ–º–µ—Ä–∞ –æ—Ç 0 –¥–æ 2)
labels = km.fit_predict(X_sc)
 
# —Å–æ—Ö—Ä–∞–Ω—è–µ–º –º–µ—Ç–∫–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞ –≤ –ø–æ–ª–µ –Ω–∞—à–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
df['cluster'] = labels
```
